## Задание 1: Фермер

**Условие:** На ферме содержатся шесть разных видов животных. Каждый раз, когда фермер заходит в сарай, он видит одно случайное животное. За день фермер заходит в сарай 6 раз. Каково математическое ожидание количества **разных** видов животных, которые фермер увидит за день?

**Решение:**

Каждый заход — равновероятный выбор одного из 6 видов. Применяем метод индикаторных случайных величин: для каждого вида $i$ вводим индикатор $X_i = 1$, если вид встретился хотя бы раз.

Вероятность того, что конкретный вид **не** попался ни разу за 6 заходов:

$$P(\text{не встретился}) = \left(\frac{5}{6}\right)^6$$

Вероятность того, что вид встретился хотя бы раз:

$$P(X_i = 1) = 1 - \left(\frac{5}{6}\right)^6 \approx 0{,}6651$$

По линейности математического ожидания:

$$E[X] = 6 \cdot \left(1 - \left(\frac{5}{6}\right)^6\right) = 6 \cdot 0{,}6651 \approx 3{,}99$$

**Ответ: 3,99**

## Задание 2: Кулинарное соревнование

**Условие:** В конкурсе участвуют 80 шеф-поваров с уникальными уровнями мастерства. В каждом поединке побеждает тот, у кого уровень мастерства выше. В первом этапе шефы случайно разбиваются на пары. Во втором этапе победители первого этапа снова случайно разбиваются на пары (повторения допускаются). Победную награду получают те, кто выиграл в **обоих** этапах. Каково математическое ожидание числа победителей?

**Решение:**

Исход каждого поединка детерминирован (побеждает сильнейший) — случайность лишь в формировании пар.

- **Этап 1:** 80 шефов → 40 пар → ровно **40 победителей** при любом разбиении.
- **Этап 2:** 40 победителей → 20 пар → ровно **20 победителей** при любом разбиении.

Число итоговых победителей **не зависит** от жеребьёвки — оно всегда равно 20. Математическое ожидание константы равно ей самой:

$$E[X] = 20$$

**Ответ: 20**

## Задание 3: Одинокая дорога

**Условие:** На пустынном шоссе вероятность появления автомобиля за 30-минутный период равна 0,95. Какова вероятность его появления за 10 минут? За 27 минут?

**Решение:**

Появление машины моделируем пуассоновским процессом с интенсивностью $\lambda$. Вероятность появления хотя бы одной машины за время $t$:

$$P(t) = 1 - e^{-\lambda t}$$

Из условия для $t = 30$ мин находим $\lambda$:

$$1 - e^{-30\lambda} = 0{,}95 \implies e^{-30\lambda} = 0{,}05 \implies \lambda = -\frac{\ln(0{,}05)}{30}$$

Тогда для произвольного $t$:

$$P(t) = 1 - 0{,}05^{\,t/30}$$

**При $t = 10$ мин:**
$$P(10) = 1 - 0{,}05^{1/3} \approx 0{,}6316 \quad (63{,}2\%)$$

**При $t = 27$ мин:**
$$P(27) = 1 - 0{,}05^{0{,}9} \approx 0{,}9327 \quad (93{,}3\%)$$

**Ответ: 63,2% и 93,3%**
