# Блок 1: Теория вероятности и логика

> ## Задание 1: Фермер
> ```
> На ферме содержатся шесть разных видов животных, и каждый раз, когда фермер заходит в сарай, он видит одно случайное животное. За день фермер заходит в сарай 6 раз.
> ```
> ```
> Каково математическое ожидание количества разных видов животных, которые фермер увидит за день?
> ```
Каждый заход — равновероятный выбор одного из 6 видов. Используем индикаторы: для каждого вида смотрим, встретился ли он хотя бы раз за 6 заходов. Итоговое ожидаемое число видов — это 6, умноженное на вероятность того, что конкретный вид встретился хотя бы раз.
Вероятность того, что вид не попался ни разу за 6 заходов — это (5/6)^6, значит вероятность обратного — 1 минус (5/6)^6, что примерно равно 0,6651.

-> E[X] = 6 * 0,6651 = 3,99
#### Ответ: 3,99

> ## Задание 2: Кулинарное соревнование
> ```
> В конкурсе участвуют 80 шеф-поваров с уникальными уровнями мастерства. В первом этапе судьи случайным образом распределяют их по парам  
> (в любом состязании двух шефов выигрывает тот, у кого выше уровень мастерства). На втором этапе шефы снова случайно образуют пары для
> финального раунда (пары могут повториться). Победную награду получают те, кто выиграл в обоих этапах.
> Каково математическое ожидание числа победителей?
> ```

80 шефов, у каждого уникальный уровень мастерства. В каждой паре всегда побеждает сильнейший — случайность только в том, кто с кем попарится.
В первом этапе 80 шефов разбиваются на 40 пар, из каждой пары выходит один победитель — итого всегда ровно 40 победителей, при любом разбиении. Во втором этапе эти 40 победителей разбиваются на 20 пар — итого всегда ровно 20 победителей.
Число итоговых победителей не зависит от того, как именно выпали пары — оно всегда равно 20. Матожидание константы — это она сама.
#### Ответ: 20,0

> ## Задание 3: Одинокая дорога
> ```
> На пустынном шоссе вероятность появления автомобиля за 30-минутный период составляет 0.95.
> Какова вероятность его появления за 10 минут? А за 27 минут?
> ```
Появление машины моделируется пуассоновским процессом. Вероятность того, что за время t не появится ни одной машины, убывает экспоненциально.
Из условия на 30 минут находим интенсивность потока: вероятность отсутствия машины равна 0,05, откуда lambda = -ln(0,05) / 30.
Для любого t вероятность появления машины считается как 1 минус 0,05 в степени t/30.
При t = 10 минут: 1 - 0,05^(1/3) = 0,6316, то есть 63,2%
При t = 27 минут: 1 - 0,05^(0,9) = 0,9327, то есть 93,3%
#### Ответ: 63,2; 93,3
