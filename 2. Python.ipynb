## Задание 1: Изоморфизм строк

**Определение:** Две строки `s` и `t` называются **изоморфными**, если существует взаимно однозначное отображение символов `s → t`, при котором сохраняется порядок символов и позиции их вхождений.

**Идея алгоритма:**  
Храним два словаря:
- `s_to_t` — отображение символов из `s` в `t`
- `t_to_s` — обратное отображение из `t` в `s`

На каждом шаге проверяем согласованность обоих отображений. Если встречаем противоречие — строки не изоморфны.

**Оценка сложности:**
| | Оценка |
|---|---|
| **Время** | O(n) — один проход по строке |
| **Память** | O(k) — где k — размер алфавита (не более 256 для ASCII) |

```python
def is_isomorphic(s: str, t: str) -> bool:
    """
    Проверяет, являются ли строки s и t изоморфными.
    
    Временная сложность: O(n)
    Пространственная сложность: O(k), k — размер алфавита
    """
    if len(s) != len(t):
        return False

    s_to_t = {}  # символ из s -> символ из t
    t_to_s = {}  # символ из t -> символ из s

    for ch_s, ch_t in zip(s, t):
        # проверяем отображение s -> t
        if ch_s in s_to_t:
            if s_to_t[ch_s] != ch_t:
                return False
        else:
            s_to_t[ch_s] = ch_t

        # Проверяем обратное отображение t -> s
        if ch_t in t_to_s:
            if t_to_s[ch_t] != ch_s:
                return False
        else:
            t_to_s[ch_t] = ch_s

    return True


# --- Тест ---
print("=" * 30)
print("Задание 1: Изоморфизм строк")
print("=" * 30)

test_cases = [
    ("paper", "title", True),
    ("egg",   "add",   True),
    ("foo",   "bar",   False),
    ("ab",    "aa",    False),
    ("badc",  "bab",  False),
]

for s, t, expected in test_cases:
    result = is_isomorphic(s, t)
    status = "✓" if result == expected else "✗"
    print(f"{status}  is_isomorphic({s!r}, {t!r}) = {result}  (ожидалось: {expected})")
```

## Задание 2: Пропущенное число в натуральной последовательности

**Определение:** Дан список из n−1 различных натуральных чисел в диапазоне от 1 до n. Требуется найти единственное отсутствующее число.

**Идея алгоритма:**  
Используем формулу суммы арифметической прогрессии:

$$S_{\text{ожидаемая}} = \frac{n \cdot (n+1)}{2}$$

Пропущенное число = $S_{\text{ожидаемая}} - S_{\text{фактическая}}$

**Оценка сложности:**
| | Оценка |
|---|---|
| **Время** | O(n) — один проход для вычисления суммы |
| **Память** | O(1) — константная, не используем вспомогательных структур |

```python
def missing_number(nums: list) -> int:
    """
    Находит единственное пропущенное число в последовательности 1..n.
    
    Алгоритм:
    - n = len(nums) + 1
    - Ожидаемая сумма = n*(n+1)//2
    - Пропущенное = ожидаемая сумма - фактическая сумма
    
    Временная сложность: O(n)
    Пространственная сложность: O(1)
    """
    n = len(nums) + 1   # полная длина последовательности
    expected_sum = n * (n + 1) // 2
    actual_sum = sum(nums)
    return expected_sum - actual_sum


# --- Тест ---
print("=" * 30)
print("Задание 2: Пропущенное число")
print("=" * 30)

test_cases = [
    ([1, 2, 3, 4, 5, 6, 8, 9, 10, 11], 7), #  S1=59; S2=66
    ([2, 3, 4, 5],                      1),
    ([1, 2, 3, 4],                      5),
    ([1],                               2),
]

for nums, expected in test_cases:
    result = missing_number(nums)
    status = "✓" if result == expected else "✗"
    print(f"{status}  missing_number({nums}) = {result}  (ожидалось: {expected})")
```

## Задание 3: Факторизация числа

**Определение:** **Факторизация** — разложение натурального числа n на простые множители, то есть представление в виде произведения простых чисел.

**Пример:** 56 = 2 × 2 × 2 × 7

**Идея алгоритма (пробное деление):**
1. Делим n последовательно на 2, 3, 4, … пока делитель d не превысит √n
2. Пока n делится на d — добавляем d в список и делим n на d
3. Если после цикла n > 1, значит оставшееся n само является простым множителем

**Оптимизация:** проверяем делители только до √n, что даёт значительное ускорение по сравнению с полным перебором.

**Оценка сложности:**
| | Оценка |
|---|---|
| **Время** | O(√n) — перебор делителей до корня из n |
| **Память** | O(log n) — количество простых множителей не превышает log₂(n) |

```python
def prime_factors(n: int) -> list:
    """
    Возвращает список простых множителей числа n в порядке возрастания.
    
    Алгоритм (пробное деление):
    - Делим на каждый делитель d от 2 до sqrt(n)
    - Если n делится на d — записываем d и продолжаем деление
    - После цикла: если n > 1, оно само является простым множителем
    
    Временная сложность: O(sqrt(n))
    Пространственная сложность: O(log n)
    """
    if n < 2:
        return []

    factors = []
    d = 2

    while d * d <= n:       # перебираем делители до sqrt(n)
        while n % d == 0:   # пока n делится на d без остатка
            factors.append(d)
            n //= d
        d += 1

    if n > 1:               # оставшийся множитель — простое число
        factors.append(n)

    return factors


# --- Тесты ---
print("=" * 40)
print("Задание 3: Факторизация числа")
print("=" * 40)

test_cases = [
    (56,  [2, 2, 2, 7]),
    (12,  [2, 2, 3]),
    (100, [2, 2, 5, 5]),
    (13,  [13]),        # простое число
    (1,   []),          # граничный случай
    (360, [2, 2, 2, 3, 3, 5]),
]

for n, expected in test_cases:
    result = prime_factors(n)
    status = "✓" if result == expected else "✗"
    print(f"{status}  prime_factors({n}) = {result}  (ожидалось: {expected})")
```
