<h2>Задание 1 — Воодушевлённое руководство</h2>
<div class='ctx'>
Команда провела А/Б-эксперимент по замене транспортного средства доставщиков в <strong>небольшом городке РФ</strong>. Результаты превзошли ожидания: время доставки значимо снизилось. Руководство хочет применить изменения по всей стране.
</div>

**Вопрос:** Делаем? Выберите все верные утверждения:



**Верные ответы: 3 и 4.**

Суть в понятии **репрезентативности**: выборка репрезентативна относительно той **генеральной совокупности**, из которой она взята. Тест проводился в одном небольшом городе — значит, его результаты обобщаются только на **похожие малые города**.

| Вариант | Почему неверен / верен |
|---|---|
| Вариант 1 | Неверен. Нельзя автоматически переносить результат на всю страну — крупные города отличаются по инфраструктуре, плотности, трафику |
| Вариант 2 | Неверен. Тест **репрезентативен** — просто его генеральная совокупность ≠ вся Россия, а только малые города |
| Вариант 3 | ✅ Верен. Именно такова область применения результатов |
| Вариант 4 | ✅ Верен. Грамотная стратегия масштабирования: сначала подтвердить эффект в других типах городов |


<div class='task'>
<h2>Задание 2 — Основной показатель в статистике: p-value</h2>

**Вопрос:** Что такое p-value в статистическом тестировании? Выберите одно верное утверждение:


p-value — это вероятность, вычисленная **при предположении, что H₀ верна**. Она показывает, насколько «удивительны» наши данные в мире, где нет никакого эффекта.

**Верный ответ: 2.**

**p-value = P(получить данные ≥ наблюдаемых | H₀ истинна)**.


| Вариант | Почему неверен |
|---|---|
| Вариант 1 | Классическое заблуждение. p-value — это P(данные \| H₀), а не P(H₀ \| данные). Второе — байесовская вероятность, которую t-тест не даёт |
| Вариант 3 | Уровень значимости α — это **порог**, задаваемый заранее. p-value сравнивается с α, но не является им |
| Вариант 4 | Среднее — просто описательная статистика, не вероятность |
